```python
%matplotlib inline
import matplotlib.pyplot as plt

# 获取output 文件中数据的epch 和对应的数据
clist = []
with open("../output_opendata_train.txt","r") as f:
    for line in f:
        if "Cosine-Similarity" in line:
            index=float(line.strip().split(":")[-1].strip())
            index = float(line.strip().split("Pearson:")[-1].split("	Spearman:")[0])
            clist.append(index)

steps = []
for i in range(len(clist)):
    steps.append((i+1)*1000)



# 定义epoch和steps
epochs = [0, 0, 1, 1, 2, 2, 3, 3]  # 每个epoch的值复制两次，以匹配每个step的数量

# 定义评估结果
spearman_scores = clist

# 创建画布和子图
fig, ax = plt.subplots()

# 绘制折线图
#ax.plot(steps, pearson_scores, label='Pearson')
ax.plot(steps, spearman_scores, label='Spearman')

# 添加标题和标签
ax.set_title('Evaluation Scores')
ax.set_xlabel('Steps')
ax.set_ylabel('Scores')

# 添加图例
ax.legend()

# 显示图形
# 
plt.show()
```
